```python
import mediapipe as mp 
import cv2
import numpy as np

mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh()
mpDraw = mp.solutions.drawing_utils
faceLmsStyle = mpDraw.DrawingSpec(color = (0, 0, 255), thickness = 1)
faceConnStyle = mpDraw.DrawingSpec(color = (0, 255, 0), thickness = 1)

FaceContourList = [10, 338, 297, 332, 284, 251, 389, 356, 454, 323, 361, 288, 397, 365, 379, 378, 400, 152, 148, 176, 149, 150, 136, 172, 58, 132, 93, 234, 127, 162, 21, 54, 103, 67, 109]
LeftEyeList = [33, 246, 161, 160, 159, 158, 157, 173, 133, 155, 154, 153, 145, 144, 163, 7]
RightEyeList = [362, 398, 384, 385, 386, 387, 388, 466, 263, 249, 390, 373, 374, 380, 381, 382]
MouthList = [13, 312, 311, 310, 415, 292, 324, 318, 402, 317, 14, 87, 178, 88, 95, 62, 191, 80, 81, 82]
NoseLeftList = [326, 327, 294, 278, 344, 440, 274, 354, 370]
NoseRightList = [97, 141, 125, 44, 220, 115, 48, 64, 240]
LeftEyebrowList = [300, 293, 334, 296, 336, 285, 295, 282, 283, 276]
RightEyebrowList = [70, 63, 105, 66, 107, 55, 65, 52, 53, 46]

cap = cv2.VideoCapture(0)

while cap.isOpened():
    # 從攝影機擷取一張影像
    ret, frame = cap.read()
    frame = cv2.cvtColor(cv2.flip(frame, 1), cv2.COLOR_BGR2RGB)
    frame.flags.writeable = False
    result = face_mesh.process(frame)
    frame.flags.writeable = True
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
    imgHeight = frame.shape[0]
    imgWidth = frame.shape[1]
    FaceContourPosList_tmp = np.array([[0, 0, 0]])
    LeftEyePosList_tmp = np.array([[0, 0, 0]])
    RightEyePosList_tmp = np.array([[0, 0, 0]])
    MouthPosList_tmp = np.array([[0, 0, 0]])
    NoseLeftPosList_tmp = np.array([[0, 0, 0]])
    NoseRightPosList_tmp = np.array([[0, 0, 0]])
    LeftEyebrowPosList_tmp = np.array([[0, 0, 0]])
    RightEyebrowPosList_tmp = np.array([[0, 0, 0]])

    FaceContourPosList = np.array([[0, 0]])
    LeftEyePosList = np.array([[0, 0]])
    RightEyePosList = np.array([[0, 0]])
    MouthPosList = np.array([[0, 0]])
    NoseLeftPosList = np.array([[0, 0]])
    NoseRightPosList = np.array([[0, 0]])
    LeftEyebrowPosList = np.array([[0, 0]])
    RightEyebrowPosList = np.array([[0, 0]])
    skin_mask = np.zeros((imgHeight, imgWidth, 3), np.uint8)
    if(result.multi_face_landmarks):
        for faceLMS in result.multi_face_landmarks:
            mpDraw.draw_landmarks(frame, faceLMS, mp_face_mesh.FACEMESH_CONTOURS, faceLmsStyle, faceConnStyle)
            for i, lm in enumerate(faceLMS.landmark):
                px = int(lm.x * imgWidth)
                py = int(lm.y * imgHeight)
                #cv2.putText(frame, str(i), (px, py), cv2.FONT_HERSHEY_SIMPLEX, 0.2, (0, 255, 0), 1)
                for num in FaceContourList:
                    if num == i:
                        FaceContourPosList_tmp = np.append(FaceContourPosList_tmp, np.array([[i, px, py]]), axis = 0)
                for num in LeftEyeList:
                    if num == i:
                        LeftEyePosList_tmp = np.append(LeftEyePosList_tmp, np.array([[i, px, py]]), axis = 0)
                for num in RightEyeList:
                    if num == i:
                        RightEyePosList_tmp = np.append(RightEyePosList_tmp, np.array([[i, px, py]]), axis = 0)
                for num in MouthList:
                    if num == i:
                        MouthPosList_tmp = np.append(MouthPosList_tmp, np.array([[i, px, py]]), axis = 0) 
                for num in NoseLeftList:
                    if num == i:
                        NoseLeftPosList_tmp = np.append(NoseLeftPosList_tmp, np.array([[i, px, py]]), axis = 0)
                for num in NoseRightList:
                    if num == i:
                        NoseRightPosList_tmp = np.append(NoseRightPosList_tmp, np.array([[i, px, py]]), axis = 0)
                for num in LeftEyebrowList:
                    if num == i:
                        LeftEyebrowPosList_tmp = np.append(LeftEyebrowPosList_tmp, np.array([[i, px, py]]), axis = 0)
                for num in RightEyebrowList:
                    if num == i:
                        RightEyebrowPosList_tmp = np.append(RightEyebrowPosList_tmp, np.array([[i, px, py]]), axis = 0)


        for num in FaceContourList:
            for [a, b, c] in FaceContourPosList_tmp:
                if num == a:
                    FaceContourPosList = np.append(FaceContourPosList, np.array([[b, c]]), axis = 0)                
        for num in LeftEyeList:
            for [a, b, c] in LeftEyePosList_tmp:
                if num == a:
                    LeftEyePosList = np.append(LeftEyePosList, np.array([[b, c]]), axis = 0)  
        for num in  RightEyeList:
            for [a, b, c] in RightEyePosList_tmp:
                if num == a:
                    RightEyePosList = np.append(RightEyePosList, np.array([[b, c]]), axis = 0)
        for num in MouthList:
            for [a, b, c] in MouthPosList_tmp:
                if num == a:
                    MouthPosList = np.append(MouthPosList, np.array([[b, c]]), axis = 0) 
        for num in NoseLeftList:
            for [a, b, c] in NoseLeftPosList_tmp:
                if num == a:
                    NoseLeftPosList = np.append(NoseLeftPosList, np.array([[b, c]]), axis = 0) 
        for num in NoseRightList:
            for [a, b, c] in NoseRightPosList_tmp:
                if num == a:
                    NoseRightPosList = np.append(NoseRightPosList, np.array([[b, c]]), axis = 0) 
        for num in LeftEyebrowList:
            for [a, b, c] in LeftEyebrowPosList_tmp:
                if num == a:
                    LeftEyebrowPosList = np.append(LeftEyebrowPosList, np.array([[b, c]]), axis = 0) 
        for num in RightEyebrowList:
            for [a, b, c] in RightEyebrowPosList_tmp:
                if num == a:
                    RightEyebrowPosList = np.append(RightEyebrowPosList, np.array([[b, c]]), axis = 0) 

        FaceContourPosList = np.delete(FaceContourPosList, np.array([[0, 0]]), axis = 0)            
        LeftEyePosList = np.delete(LeftEyePosList, np.array([[0, 0]]), axis = 0)  
        RightEyePosList = np.delete(RightEyePosList, np.array([[0, 0]]), axis = 0)
        MouthPosList = np.delete(MouthPosList, np.array([[0, 0]]), axis = 0)
        NoseLeftPosList = np.delete(NoseLeftPosList, np.array([[0, 0]]), axis = 0)
        NoseRightPosList = np.delete(NoseRightPosList, np.array([[0, 0]]), axis = 0)
        LeftEyebrowPosList = np.delete(LeftEyebrowPosList, np.array([[0, 0]]), axis = 0)
        RightEyebrowPosList = np.delete(RightEyebrowPosList, np.array([[0, 0]]), axis = 0)

        cv2.fillPoly(skin_mask, [FaceContourPosList], (255, 255, 255))
        cv2.fillPoly(skin_mask, [LeftEyePosList], (0, 0, 0))
        cv2.fillPoly(skin_mask, [RightEyePosList], (0, 0, 0))
        cv2.fillPoly(skin_mask, [MouthPosList], (0, 0, 255))
        cv2.fillPoly(skin_mask, [NoseLeftPosList], (0, 0, 0))
        cv2.fillPoly(skin_mask, [NoseRightPosList], (0, 0, 0))
        cv2.fillPoly(skin_mask, [LeftEyebrowPosList], (0, 0, 0))
        cv2.fillPoly(skin_mask, [RightEyebrowPosList], (0, 0, 0))
        cv2.imshow('skin',  skin_mask)
        cv2.waitKey(1)
    # 顯示圖片
    cv2.imshow('frame', frame)
    cv2.waitKey(1)
    # 若按下 q 鍵則離開迴圈
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 釋放攝影機
cap.release()

# 關閉所有 OpenCV 視窗
cv2.destroyAllWindows()
```
